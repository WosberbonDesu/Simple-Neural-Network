# Machine Learning Part

```python
# Lets define our x and y value for machine learning 
# We will prepare a model according to its type and train it

y = df["Type"].values
x = df.drop("Type",axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
# we will allocate thirty percent for the test value
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

## Neural Network Part


```python
model = Sequential()
# we will use relu for activation its better then others and the first will have 30 neurons
# Optimizer  will be adam its also better then other

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1)
# 700 epochs enough i think
# There is no need to redefine the validation_data, we can also give it in the model
# Verbose 1 its okay it will show us the loss value one by one
```

```python
# let's see the list
model.history.history
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python

modelLoss.plot();
# we expect these two very bad values ​​to be very close to each other, but they are
# this means our neurons are not making accurate predictions

```

### EarlyStopping Tool

```python
# We will use our early stop tool
# will stop automatically after they start to be far from each other
# it doesn't need to be epoch 700

# Lets define neural network again
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
# using this we will provide what I said above
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
# using callbacks we can define earlystopping parameter 
```

```python
# Let's observe the model again and make sure it works.
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot();
# looks fine but not very accurate still some error

```

## DropOut  Tool

```python
# we define this parameter under each neural network
# This is doing the extraction of sixty percent of the incoming data, in fact over fifty percent can corrupt the data a little
# But don't worry, our data is pretty solid

model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
lossDf = pd.DataFrame(model.history.history)
```

```python
lossDf.plot();
# slightly better than before

#zigzags are visible as data has been removed
```

```python
# now we can read our test values ​​to our neurons
# In this way, we can calculate mathematically how many percent accuracy our model has.
ouprrr = model.predict_classes(x_test)
```

```python
ouprrr
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,ouprrr))
# this great model has an accuracy of near precision 97%

```

```python
print(confusion_matrix(y_test,ouprrr))
```
